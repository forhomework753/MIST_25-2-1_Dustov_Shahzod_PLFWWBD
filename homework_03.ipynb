```python
def count_word_frequency(text):
    text = text.split(" ")
    punctuationSymbols = [",",".","!","?"]
    wordFrequency = {}
    for idx, word in enumerate(text):
        for punctuation in punctuationSymbols:
            word = word.replace(punctuation,"")
        text[idx] = word.lower()
    for word in text:
        count = 0
        for currentWord in text:
            if word == currentWord:
                count+=1
        wordFrequency[word] = count
    return wordFrequency
print(count_word_frequency("Hello, its me DS. How are you? Hello DS I am fine, how are you?"))
```

```python
def withGetMethod(text):
    text = text.split(" ")
    punctuationSymbols = [",",".","!","?"]
    wordFrequency = {}
    for idx, word in enumerate(text):
        for punctuation in punctuationSymbols:
            word = word.replace(punctuation,"")
        word = word.lower()
        wordFrequency[word] = wordFrequency.get(word,0)+1
    return wordFrequency
print(withGetMethod("Hello, its me DS. How are you? Hello DS I am fine, how are you?"))
```

```python
def calculate_price(price,discount=0,tax_rate=0.2):
    priceWithDiscount = price - price*discount
    productPriceAfterTaxes = priceWithDiscount + priceWithDiscount * tax_rate
    return productPriceAfterTaxes
calculate_price(100,0.5)
```

```python
import datetime
def analyze_log(file_path):
    result = {"counts":{}}
    try:
        logs = open(file_path,"r+")
        logDates = []
        max_date = 0
        min_date = 0
        for log in logs:
            try:
                log = log.split(",")
                logDates.append(datetime.datetime.strptime(log[0],"%Y-%m-%d %H:%M:%S"))
                logLevel = log[1]
                min_date = logDates[0]
                max_date = logDates[0]
                result["counts"][logLevel] = result["counts"].get(logLevel,0)+1
            except (IndexError, ValueError):
                continue
        for currentDate in logDates:
            if currentDate < min_date:
                min_date = currentDate
            if currentDate > max_date:
                max_date = currentDate
        result["time_range"] = (min_date.strftime("%Y-%m-%d %H:%M:%S"), max_date.strftime("%Y-%m-%d %H:%M:%S"))
        return result
    except FileNotFoundError:
        print(f"[-] Файл: {file_path} не найден")
        return {}
    finally:
        logs.close()
analyze_log("./log.txt")
```
