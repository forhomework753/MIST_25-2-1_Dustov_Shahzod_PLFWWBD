# Задание 1

```python
while True:
    print("[!] Для выхода напишите число 2004")
    number = int(input("[+] Введите целое положительное число: "))
    if number < 0:
        print("[!] Число не положительная, введите заново")
        continue
    if number == 2004:break
    for fizz in range(1, number+1):
        if fizz % 3 == 0 and fizz % 5 == 0:
            print("FizzBuzz")
        elif fizz % 3 == 0:
            print("Fizz")
        elif fizz % 5 == 0:
            print("Buzz")
        else:
            print(fizz)
```

# Задание 2

```python
shopping_list = []
while True:
    print("1> Добавить 2> Показать 3> Выйти")
    selected = input("[?] Выберите действие(число): ") 
    if selected == "1":
        productName = input("[~] Какой продукт хочете добавить в список?: ")
        shopping_list.append(productName)
        print("[+] Продукт успешно добавлен в список")
    elif selected == "2":
        if not len(shopping_list):
            print("[!] Список покупок пуст.")
            continue
        for idx, product in enumerate(shopping_list):
            print(f"{idx+1}.{product}")
    elif selected == "3": 
        print("[!] Программа завершена")
        break
    else:
        print("[!] Неизвестная команда, попробуйте снова")
            
```
